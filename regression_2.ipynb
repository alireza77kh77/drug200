```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('drug200.csv')
dataset
```

```python
df = pd.DataFrame(dataset)
df
```

```python
df.drop(['Age'],axis=1, inplace=True)
```

```python
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()
df['BP']= label_encoder.fit_transform(df['BP'])

df['BP'].unique()
```

```python
df['Drug']= label_encoder.fit_transform(df['Drug'])

df['Drug'].unique()
```

```python
df['Cholesterol']= label_encoder.fit_transform(df['Cholesterol'])

df['Cholesterol'].unique()
```

```python
df
```

```python
dataset.shape
```

```python
def min_max_normallize(dataset):
    min_val = np.min(dataset, axis=0)
    max_val = np.max(dataset, axis=0)
    normal = (dataset - min_val) / (max_val - min_val)
    return normal
```

```python
df = min_max_normallize(df)
df
```

```python
x = df.iloc[:, :5]
y = df.iloc[:, 0]
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print('MSE=', mse)
print('R2=', r2)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree=8)
x_poly = poly_reg.fit_transform(x_test)
model.fit(x_poly, y_test)
```

```python
y_pred = model.predict(x_poly)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print('MSE=', mse)
print('R2=', r2)
```
